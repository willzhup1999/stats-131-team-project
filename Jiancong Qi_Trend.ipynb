```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
crime = pd.read_csv("/Users/jiancongqi/OneDrive/UCLA/Fall 2019/Statistics 131/crime-in-vancouver/crime.csv")

Break_Commercial = crime[crime['TYPE'] == "Break and Enter Commercial"]
Break_ResidentialOther = crime[crime['TYPE'] == "Break and Enter Residential/Other"]
Homicide = crime[crime['TYPE'] == "Homicide"]
Mischief = crime[crime['TYPE'] == "Mischief"]
Offence = crime[crime['TYPE'] == "Offence Against a Person"]
Other_Theft = crime[crime['TYPE'] == "Other Theft"]
Theft_Vehicle = crime[crime['TYPE'] == "Theft from Vehicle"]
Theft_Bicycle = crime[crime['TYPE'] == "Theft of Bicycle"]
Collision_Fatality = crime[crime['TYPE'] == "Vehicle Collision or Pedestrian Struck (with Fatality)"]
Collision_Injury = crime[crime['TYPE'] == "Vehicle Collision or Pedestrian Struck (with Injury)"]
```

```python

#Mischief:
#Year: A huge outlier in June 2011. Decline over years.
#Month: No specific trend
#Day: No specific trend
#Hour: No specific trend
new = pd.crosstab(index=Mischief.YEAR, columns=Mischief.MONTH)
stacked = new.stack().reset_index().rename(columns={0:'value'})
sns.barplot(x=stacked.MONTH, y=stacked.value, hue=stacked.YEAR).legend(loc='center left', bbox_to_anchor=(1.25, 0.5), ncol=1)
```

```python

#Vehicle theft
#Year: U-shape. 2011-2013 have the lowest number of Vehicle theft.
#Month: No specific trend
#Day: No specific trend
#Hour: No specific trend

new = pd.crosstab(index=Theft_Vehicle.MONTH, columns=Theft_Vehicle.YEAR)
stacked = new.stack().reset_index().rename(columns={0:'value'})
sns.barplot(x=stacked.YEAR, y=stacked.value, hue=stacked.MONTH).legend(loc='center left', bbox_to_anchor=(1.25, 0.5), ncol=1)
```
